1. 從Kaggle擷取資料, 初步檢視資料
----

```python
# import 所需 lib
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import preprocessing, linear_model
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

url = "https://storage.googleapis.com/2017_ithome_ironman/data/kaggle_titanic_train.csv"
titanic_train = pd.read_csv(url)

#初步檢視資料
titanic_train.head()
```

```python
titanic_train.describe()
```

```python
#修補 Age NaN問題, 以 medium 修補缺少資料
age_median = np.nanmedian(titanic_train["Age"])
new_Age = np.where(titanic_train["Age"].isnull(), age_median, titanic_train["Age"])
titanic_train["Age"] = new_Age
titanic_train.describe()
```

```python
titanic_train.dtypes
```

2. 試著用分類樹建立模型, 並測試績效
----

```python
from sklearn import tree
from sklearn.cross_validation import train_test_split
from sklearn import metrics

# 切分 train / test 資料
data_X = pd.DataFrame([titanic_train["Age"],
                    titanic_train["Pclass"], 
                    titanic_train["SibSp"], 
                    titanic_train["Parch"], 
                    titanic_train["Fare"]]).T

data_Y = pd.DataFrame([titanic_train["Survived"]]).T

train_X, test_X, train_y, test_y = train_test_split(data_X, data_Y, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
titanic_clf = clf.fit(train_X, train_y)

# 預測
test_y_predicted = titanic_clf.predict(test_X)
print(test_y_predicted)

# 績效
accuracy = metrics.accuracy_score(test_y, test_y_predicted)
print(accuracy)
```

> 分類樹準確率約60% ~ 66%
