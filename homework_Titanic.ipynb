1. 從Kaggle擷取資料, 初步檢視資料
----

```python
# import 所需 lib
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import preprocessing, linear_model
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

url = "https://storage.googleapis.com/2017_ithome_ironman/data/kaggle_titanic_train.csv"
titanic_train = pd.read_csv(url)

#初步檢視資料
titanic_train.head()
```

```python
titanic_train.describe()
```

```python
#修補 Age NaN問題, 以 medium 修補缺少資料, 並以1代表male, 2代表female
age_median = np.nanmedian(titanic_train["Age"])
new_Age = np.where(titanic_train["Age"].isnull(), age_median, titanic_train["Age"])
titanic_train["Age"] = new_Age
new_Sex = np.where(titanic_train["Sex"]=="male", 1, 2)
titanic_train["Sex"] = new_Sex
titanic_train.describe()
```

```python
titanic_train.dtypes
```

```python
# Drop Ticket and Cabin which have too many NaN
# Basic Exploratory Visualization
titanic_train = titanic_train.drop(['Ticket','Cabin'], axis=1)
sns.pairplot(titanic_train)
```

2. 試著用分類樹建立模型, 並測試績效
----

```python
from sklearn import tree
from sklearn.cross_validation import train_test_split
from sklearn import metrics

# 切分 train / test 資料
data_X = pd.DataFrame([titanic_train["Age"],
                    titanic_train["Sex"],
                    titanic_train["Pclass"], 
                    titanic_train["SibSp"], 
                    titanic_train["Parch"], 
                    titanic_train["PassengerId"],
                    titanic_train["Fare"]]).T

data_Y = pd.DataFrame([titanic_train["Survived"]]).T

train_X, test_X, train_Y, test_Y = train_test_split(data_X, data_Y, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
titanic_clf = clf.fit(train_X, train_Y)

# 預測
test_Y_predicted = titanic_clf.predict(test_X)
print(test_Y_predicted)

# 績效
accuracy = metrics.accuracy_score(test_Y, test_Y_predicted)
print(accuracy)
```

> 分類樹準確率約75%, 聽說 benchmark 是 82%

```python
# 切分 train / test 資料, 只採取 Age, Pclass 
data_X = pd.DataFrame([titanic_train["Age"],
                    titanic_train["Sex"]]).T

data_Y = pd.DataFrame([titanic_train["Survived"]]).T

train_X, test_X, train_Y, test_Y = train_test_split(data_X, data_Y, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
titanic_clf = clf.fit(train_X, train_Y)

# 預測
test_Y_predicted = titanic_clf.predict(test_X)
print(test_Y_predicted)

# 績效
accuracy = metrics.accuracy_score(test_Y, test_Y_predicted)
print(accuracy)
```

>只採取Age跟Sex的分類樹預測accuracy依然只有77%左右

3. 試著使用Random Forest Model
----
>先承認我參考了 https://github.com/agconti/kaggle-titanic

```python
import sklearn.ensemble as ske
from patsy import dmatrices
import statsmodels.api as sm

formula_ml = 'Survived ~ C(Pclass) + C(Sex) + Age + SibSp + Parch + Fare'

# Create the random forest model and fit the model to our training data
y, x = dmatrices(formula_ml, data=titanic_train, return_type='dataframe')
# RandomForestClassifier expects a 1 demensional NumPy array, so we convert
y = np.asarray(y).ravel()
#instantiate and fit our model
results_rf = ske.RandomForestClassifier(n_estimators=100).fit(x, y)

# Score the results
score = results_rf.score(x, y)
print(score)
```
